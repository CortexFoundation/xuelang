## 创建文件夹

* data：2022
    * normal :1316
    * abnormal :706

```python
from tqdm import tqdm
from glob import glob
import shutil
import os

newpath = 'data' 
if not os.path.exists(newpath):
    os.makedirs(newpath)
    
newpath_train = ['data/normal', 'data/abnormal'] 
for i in newpath_train:
    if not os.path.exists(i):
        os.makedirs(i) 
```

```python
folder_name = ['xuelang_round1_train_part1_20180628',
               'xuelang_round1_train_part2_20180705',
               'xuelang_round1_train_part3_20180709']
```

```python
for i in folder_name:
    with tqdm(glob(i+'/*/*.jpg')) as pbar:
        for fname in pbar:
            if '正常' in fname:
                shutil.copy(fname, 'data/normal')
            else:
                shutil.copy(fname, 'data/abnormal')
```
