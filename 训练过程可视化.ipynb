```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('logs/irv2_0802_f2.5.csv')[6:].reset_index(drop=True)

plt.figure(figsize=(12, 6))
ax = plt.subplot(1, 2, 1)
df['val_auc'].plot(ax=ax)

p = np.where(df['lr'][:-1].values != df['lr'][1:].values)[0]
plt.scatter(x=p, y=df['val_auc'][p].values)

ax = plt.subplot(1, 2, 2)
df[['acc', 'val_acc']].plot(ax=ax)
p = np.where(df['lr'][:-1].values != df['lr'][1:].values)[0]
plt.scatter(x=p, y=df['val_acc'][p].values)
```
